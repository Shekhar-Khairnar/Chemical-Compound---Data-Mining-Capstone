```python
import os

file_locations_sdf = []

sdf_directory = r'C:/Users/shekh/Desktop/Data mining project files/sdf'
for entry in os.scandir(sdf_directory):
    if (entry.path.endswith(".sdf")):
            #or entry.path.endswith(".png")) and entry.is_file():
        file_locations_sdf.append(entry.path)
        
        
(file_locations_sdf)#[-1]#.shape

from os import listdir
from os.path import isfile, join
file_names_sdf = [f for f in listdir(sdf_directory) if isfile(join(sdf_directory, f))]

# np.array(file_names_sdf)#.shape
```

```python
df_sdf = pd.DataFrame()

for file_path in file_locations_sdf:
    sdfFile = os.path.join(RDConfig.RDDataDir, file_path)
    
    frame = PandasTools.LoadSDF(sdfFile,smilesName='SMILES',molColName='Molecule',
               includeFingerprints=True)
    
    df_sdf = df_sdf.append(frame, ignore_index = True)
    
    #print(frame) #.columns
df_sdf.head()
```

```python
from rdkit.Chem.Descriptors import ExactMolWt
from rdkit.Chem.Crippen import MolLogP
from rdkit.Chem.rdMolDescriptors import CalcTPSA
from rdkit import Chem
```

```python
MolWt=[]
MolLogp=[]
CalTPSA=[]
for i in df_sdf['SMILES']:
    m=Chem.MolFromSmiles(i)
    MolWt.append(ExactMolWt(m))
    MolLogp.append(MolLogP(m))
    CalTPSA.append(CalcTPSA(m))
    
df_sdf['Mol_Wt']=MolWt
df_sdf['Mol_log_P']=MolLogp
df_sdf['Calc_TPSA']=CalTPSA
df_sdf.head()
```

```python
from rdkit.Chem.AtomPairs import Pairs

ms = [Chem.MolFromSmiles(i) for i in df_sdf['SMILES']]
pair_vec = [Pairs.GetAtomPairFingerprintAsBitVect(x) for x in ms]
pair_vect_binary = [i.ToBitString() for i in pair_vec]
```

```python
fps = []
for a in pair_vect_binary:
    fps1 = [ int(i) for i in a]
    fps.append((fps1))
type(fps[0])
```

```python
df_sdf['encoded_fps'] = fps
df_sdf.head(1)
```

```python

X = df_sdf['encoded_fps']
y = df_sdf[['Mol_Wt', 'Mol_log_P', 'Calc_TPSA']]

X_in = []
for i in range(len(df_sdf)):
    X_in.append(np.array(X[i]))

X_1 = np.array(X_in)

y_1 = np.array(y)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
scaler = StandardScaler()


data_size = 100
X = X_1[:data_size]
y = y_1[:data_size]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
from sklearn.decomposition import PCA
pca2 = PCA(n_components=5, random_state = 42)
pca2.fit(X_train)
X_N_dims_train = pca2.transform(X_train)
X_N_dims_train
X_N_dims_test = pca2.transform(X_test)
```

```python
X_N_dims_train.shape
```

```python
model = LinearRegression()

model.fit(X_N_dims_train, y_train)

y_pred = model.predict(X_N_dims_test)

from sklearn.metrics import r2_score
print("r2 score:", r2_score(y_test, y_pred))

p = 5
n = len(y_test)
numerator = n - 1
denominator = n - p - 1
r2_adjusted = 1 - ( (1-(r2_score(y_test, y_pred))) *(numerator/denominator))
print("Adjusted R2 score:", r2_adjusted)

from sklearn.metrics import mean_squared_error

rmse = mean_squared_error(y_test, y_pred, squared=True) 
print("Root Mean Squared Error:", rmse)

from sklearn.metrics import mean_absolute_error 
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)

linear_regression = {}
linear_regression['R2_score'] = r2_score(y_test, y_pred)
linear_regression['Adjusted_R2_score'] = r2_adjusted
linear_regression['RMSE'] = rmse
linear_regression['MAE'] = mae
```
